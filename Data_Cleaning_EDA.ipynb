```python
def convert_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%m/%d/%Y', errors='coerce').strftime('%m/%d/%Y')
    except ValueError:
        try:
            return pd.to_datetime(date_str, format='%d-%m-%Y', errors='coerce').strftime('%m/%d/%Y')
        except ValueError:
            return date_str 
customers['Birthday'] = customers['Birthday'].apply(convert_date)
```

```python
customers['State Code'] = customers['State Code'].fillna("Not available") 
```

```python
customers['Birthday'] = pd.to_datetime(customers['Birthday'], format='%m/%d/%Y', errors='coerce').dt.date
```

```python
def clean_columns(df):
    df.columns = [c.replace(' ', '_').lower() for c in df.columns]

# Applying to all dataframes
clean_columns(customers)
clean_columns(products)
clean_columns(exchangerates)
clean_columns(stores)
clean_columns(sales)
```

```python
from urllib.parse import quote_plus
import pandas as pd
from sqlalchemy import create_engine

# Define the database connection parameters
host = "localhost"
port = "3306"
username = "root"
password = "Your MySQL Password" 
database = "capstone_2"

# URL encode the password to handle special characters like '@'
encoded_password = quote_plus(password)  

# Define the engine string with the encoded password
engine_string = f"mysql+pymysql://{username}:{encoded_password}@{host}:{port}/{database}"

# Create the SQLAlchemy engine
try:
    engine = create_engine(engine_string)
    print("Connection to the database was successful.")
    
    # Ensure that the DataFrames are defined
    
    table_name = "customers"
    customers.to_sql(table_name, con=engine, if_exists='replace', index=False)
    print(f"Successfully imported '{table_name}' to SQL")

    table_name = "products"
    products.to_sql(table_name, engine, if_exists='replace', index=False)
    print(f"Successfully imported '{table_name}' to SQL")

    table_name = "exchangerates"
    exchangerates.to_sql(table_name, engine, if_exists='replace', index=False)
    print(f"Successfully imported '{table_name}' to SQL")

    table_name = "stores"
    stores.to_sql(table_name, engine, if_exists='replace', index=False)
    print(f"Successfully imported '{table_name}' to SQL")

    table_name = "sales"
    sales.to_sql(table_name, engine, if_exists='replace', index=False)
    print(f"Successfully imported '{table_name}' to SQL")

except Exception as e:
    print(f"An error occurred: {e}")

```
